```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
```

```python
german_credit_file = 'GermanCredit.csv'
germ = pd.read_csv(german_credit_file)
```

```python
germ
```

```python
#CHECK: code only checks for 'none' values, doesn't take into consideration "non-zero" specification in assignment.
#Columns dropped: other_parties, other_payment_plans, own_telephone
def check_nones(n=3):
    num_of_nones = {}
    for col in germ.columns:
        num_of_nones[col] = germ[col].tolist().count('none')
    sorted_cols = sorted(num_of_nones.items(), key=lambda x: (-x[1], list(germ.columns).index(x[0])))
    cols_to_drop = [col for col, count in sorted_cols[:n]]
    print(f"Dropped columns: {cols_to_drop}")
    return germ.drop(cols_to_drop, axis=1)

germ_dropped = check_nones(3)
print(f"Remaining columns: {germ_dropped.columns.tolist()}")
```

```python
for col in germ.columns:
    germ[col] = germ[col].apply(lambda x: x.replace("'", "") if isinstance(x, str) else x)
```

```python
checking_mapping = {"no checking": "No Checking", "<0": "Low", "0<=X<200": "Medium", ">=200": "High"}
germ["checking_status"] = germ["checking_status"].map(checking_mapping)
```

```python
savings_mapping = {"no known savings": "No Savings", "<100": "Low", "100<=X<500": "Medium", "500<=X<1000": "High", ">=1000": "High"}
germ["savings_status"] = germ["savings_status"].map(savings_mapping)
```

```python
class_mapping = {"good": "1", "bad": "0"}
germ["class"] = germ["class"].map(class_mapping)
```

```python
#CHECK: Assumed that the order "Amateur, Professional, Experienced, Expert" is from smallest to largest.
employment_mapping = {"unemployed": "Unemployed", "<1": "Amateur", "1<=X<4": "Professional", "4<=X<7": "Experienced", ">=7": "Expert"}
germ["employment"] = germ["employment"].map(employment_mapping)
```

```python
pd.crosstab(germ["foreign_worker"], germ["class"], margins=True, margins_name="Total")
```

```python
pd.crosstab(germ["employment"], germ["savings_status"], margins=True, margins_name="Count")
```

```python
#CHECK: Returning nan
avg_credit_amount = germ[(germ["personal_status"] == "male single") & (germ["employment"] == "4<=X<7")]["credit_amount"].mean()
print("Average credit amount:", avg_credit_amount)
```

```python
job_duration_avg = germ.groupby("job")["duration"].mean()
job_duration_avg
```

```python
most_common_checking = germ[germ['purpose'] == 'education']['checking_status'].value_counts().idxmax()
most_common_savings = germ[germ['purpose'] == 'education']['savings_status'].value_counts().idxmax()
print(f"Most common checking status: {most_common_checking}")
print(f"Most common savings status: {most_common_savings}")
```

```python
import matplotlib.pyplot as plt
savings_vs_personal = germ.groupby(["savings_status", "personal_status"]).size().unstack()
checking_vs_personal = germ.groupby(["checking_status", "personal_status"]).size().unstack()
fig, axes = plt.subplots(1, 2, figsize=(15, 6), sharey=True)
savings_vs_personal.plot(kind="bar", ax=axes[0], color=plt.cm.Paired.colors)
axes[0].set_title("Savings Status vs Personal Status")
axes[0].set_xlabel("Savings Status")
axes[0].set_ylabel("Number of People")
axes[0].legend(title="Personal Status")

checking_vs_personal.plot(kind="bar", ax=axes[1], color=plt.cm.Paired.colors)
axes[1].set_title("Checking Status vs Personal Status")
axes[1].set_xlabel("Checking Status")
axes[1].legend(title="Personal Status")

plt.tight_layout()
plt.show()
```

```python
filtered_data = germ[germ["credit_amount"] > 4000]
avg_age_by_property = filtered_data.groupby("property_magnitude")["age"].mean()
avg_age_by_property.plot(kind="bar", color="skyblue", figsize=(8, 6))

plt.title("Average Customer Age by Property Magnitude (Credit Amount > 4000)")
plt.xlabel("Property Magnitude")
plt.ylabel("Average Age")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
filtered_data = germ[(germ["savings_status"] == "High") & (germ["age"] > 40)]
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
filtered_data["personal_status"].value_counts().plot.pie(
    ax=axes[0], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors
)
axes[0].set_title("Personal Status")
axes[0].set_ylabel("")  


filtered_data["credit_history"].value_counts().plot.pie(
    ax=axes[1], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors
)
axes[1].set_title("Credit History")
axes[1].set_ylabel("")  

filtered_data["job"].value_counts().plot.pie(
    ax=axes[2], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors
)
axes[2].set_title("Job")
axes[2].set_ylabel("")  

plt.tight_layout()
plt.show()
```
